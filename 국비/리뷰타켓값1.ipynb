```python
import pandas as pd
df=pd.read_excel('total_list.xlsx')
df['Unnamed: 1'][0]

name_list=[]
for i in range(len(df['Unnamed: 1'])):
    name_list.append(df['Unnamed: 1'][i])

name_list
```

```python
print(name_list[82],name_list[169])
```

```python
df.rename(columns={'Unnamed: 1':'가게'},inplace=True)
df
```

```python
df_new=df.drop('Unnamed: 0',axis=1)
df_new
df_T=df_new.T
df_T.columns=name_list
df_T
```

```python
df_T.drop('가게',axis=0,inplace=True)
df_T
```

```python
visit_count=[]
for name in name_list:
    try:
        visit_count.append((df_T[name]>0).sum())
    except:
        visit_count.append(0)

len(visit_count) # name_list에서 각 이름에 대해 해당하는 열(column)에서 0보다 큰 값들의 개수를 세는 것을 목적으로 합니다.
```

```python
two_time_visit=[]
for name in name_list:
    try:
        two_time_visit.append((df_T[name]>=2).sum())
    except:
        two_time_visit.append(0)

len(two_time_visit) # name_list에서 각 이름에 대해 해당하는 열(column)에서 2 이상인 값들의 개수를 세는 것을 목적으로 합니다.
```

```python
ratio_list2=[]
for i in range(184):
    ratio_list2.append(two_time_visit[i]/visit_count[i])
    #  two_time_visit 리스트와 visit_count 리스트의 같은 인덱스에 위치한 값들을 사용하여 비율을 계산하고,
    #    그 결과를 ratio_list2 리스트에 추가하는 것을 목적으로 합니다.
```

```python
import matplotlib.pyplot as plt
plt.hist(ratio_list2)
# ratio_list2의 값들이 x축에 표시되고, 해당 값들의 빈도수가 y축에 표시됩니다.
#  이를 통해 ratio_list2의 값들이 어떤 분포를 가지고 있는지 쉽게 파악할 수 있습니다.
```

```python
two_time_visit
```

```python
two_df=pd.DataFrame(ratio_list2)
two_df.fillna(0)
two_df[0].mean()
# ratio_list2를 기반으로한 데이터프레임 two_df가 생성되고, 해당 데이터프레임의 첫 번째 열에 대한 평균값이 반환됩니다.
```
